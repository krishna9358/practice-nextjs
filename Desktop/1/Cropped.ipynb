```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/leaf4.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,thresh = cv2.threshold(gray,1,255,cv2.THRESH_BINARY)

contours,hierarchy = cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cnt = contours[0]
x,y,w,h = cv2.boundingRect(cnt)

crop = img[y:y+h,x:x+w]
cv2_imshow(crop)
cv2.imwrite('/content/sample_data/cropped.png',crop)
```

```
from PIL import Image
import matplotlib.pyplot as plt

cropped= crop[30:130, 70:150]

plt.figure(figsize=(4,5))
res_img = cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)

pixels = np.array(res_img)
plt.imshow(pixels)
plt.show()

cv2.imwrite('/content/sample_data/cropped.png', res_img)
```
